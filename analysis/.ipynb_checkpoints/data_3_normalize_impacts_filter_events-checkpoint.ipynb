```python
import xarray as xr
import numpy as np
import geopandas as gpd
import pandas as pd
```

This notebook is for normalizing the human losses and financial damages impact data of the geolocated events.

1) Population data: global population data is used at 5 years interval from 1990 to 2020. For each event, the sum of the population at the affected geometry is used to normalize the data and calculate the fraction of affected people. In some cases, the fraction exceeds 1, in this case, it is manually capped to 0.999.

2) Same is done for global GDP data. Except capping max losses, as they can exceed region's GDP

### 1) Calculate population and GDP in every geometry corresponding to disaster event

```python
def get_population(event):
    print(event)
    disasters_1 = disasters[disasters.index == event]
    pop_dis = population_data.sel(band = disasters_1.year.values[0]).rio.clip(disasters_1.geometry,disasters.crs,all_touched=True)
    return {event:pop_dis.band_data.sum().values}
```

```python
def get_gdp(event):
    print(event)
    disasters_1 = disasters[disasters.index == event]
    gdp_dis = gdp_data.sel(time = disasters_1.year.values[0]).rio.clip(disasters_1.geometry,disasters.crs,all_touched=True)
    return({event:gdp_dis.GDP_PPP.sum().values})
```

```python
events = disasters.index
```

```python
%%time
results = ThreadPool(20).map(get_population, events)
```

The main problem here is the ability to distinguish between floods and storms (most numerous types of events).
In many cases, the reporting of the event timing puts the climate variables out of sync with the reported event as a whole.

For this, we need to select the events with the highest possible accuracy.

we will use windspeed for stroms, and total precipitaion and total runoff for floods. If there is no value in the reported time window of the event (03 days before and after) that is higher than the 90th percentile, then the event is removed as no extreme values of the variables are found, and therefore linking the climate and impact data would not be possible.

```python
#select the indices of each types of events
```

```python
def check_event_validity(clim_data_event, indices,variable):
    extreme_events = []
    for ev in indices:
        bool_test = any(clim_data_event.sel(units='anomalies', landcover='all').sel(event=ev)[variable].sel(time=slice(disasters.loc[[ev]].start_day[0]-5,disasters.loc[[ev]].end_day[0]+5)).compute().data >= clim_data_event.sel(units='anomalies', landcover='all').sel(event=ev)[variable].compute().quantile(0.95).data)
        if bool_test == True:
            extreme_events.append(ev)
        else:
            pass
    return extreme_events
```
