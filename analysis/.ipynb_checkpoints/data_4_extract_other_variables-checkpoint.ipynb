```python
import xarray as xr
import numpy as np
import geopandas as gpd
import pandas as pd
```

```python
import multiprocessing
```

Extract climate variables in the 3 dimensions

```python
disasters_df = pd.read_csv("/net/projects/xaida/project_data/xaida_data_base/V3.1/disaster_database_filtered_events/disasters_df.csv")
```

```python
disasters_df = disasters_df.set_index('Dis No')
```

```python
clim_data = xr.open_zarr('/net/projects/xaida/project_data/xaida_data_base/V3.1/disaster_database_filtered_events/climate_data.zarr/')
```

```python
clim_data
```

1) Extract anomalies in the original units

```python
def summarize_event_max_anomalies_originals_duringevent(x):
    xev = disasters_df.loc[[x]]
    res = clim_data.sel(units='anomalies',event=xev.disaster_number_country.values,time=slice(xev.start_day.values[0]-2,xev.end_day.values[0]+2))[['t2m','t2mmax','t2mmin','sktmax','windgustmean','windgustmax','tccmean','spmean','rh','smsurf','tp','tro','spei_30','spei_90','spei_180']].groupby('landcover').max('time').compute().to_dataframe()
    res_df = pd.DataFrame(res.reset_index().set_index('event').query("landcover == 'all'")[['t2m','t2mmax','t2mmin','sktmax','windgustmean','windgustmax','tccmean','spmean','rh','smsurf','tp','tro','spei_30','spei_90','spei_180']])
    return(res_df)
```

```python
def summarize_event_min_anomalies_originals_duringevent(x):
    xev = disasters_df.loc[[x]]
    res = clim_data.sel(units='anomalies',event=xev.disaster_number_country.values,time=slice(xev.start_day.values[0]-2,xev.end_day.values[0]+2))[['t2m','t2mmax','t2mmin','sktmax','windgustmean','windgustmax','tccmean','spmean','rh','smsurf','tp','tro','spei_30','spei_90','spei_180']].groupby('landcover').min('time').compute().to_dataframe()
    res_df = pd.DataFrame(res.reset_index().set_index('event').query("landcover == 'all'")[['t2m','t2mmax','t2mmin','sktmax','windgustmean','windgustmax','tccmean','spmean','rh','smsurf','tp','tro','spei_30','spei_90','spei_180']])
    return(res_df)
```

```python
def summarize_event_mean_anomalies_originals_duringevent(x):
    xev = disasters_df.loc[[x]]
    res = clim_data.sel(units='anomalies',event=xev.disaster_number_country.values,time=slice(xev.start_day.values[0]-2,xev.end_day.values[0]+2))[['t2m','t2mmax','t2mmin','sktmax','windgustmean','windgustmax','tccmean','spmean','rh','smsurf','tp','tro','spei_30','spei_90','spei_180']].groupby('landcover').mean('time').compute().to_dataframe()
    res_df = pd.DataFrame(res.reset_index().set_index('event').query("landcover == 'all'")[['t2m','t2mmax','t2mmin','sktmax','windgustmean','windgustmax','tccmean','spmean','rh','smsurf','tp','tro','spei_30','spei_90','spei_180']])
    return(res_df)
```

```python
events_list = disasters_df.index
```

```python
%%time
with multiprocessing.Pool(30) as p:
    res_during_max = p.map(summarize_event_max_anomalies_originals_duringevent, events_list)
```

```python
max_anomalies_during = pd.concat(res_during_max)
```

```python
max_anomalies_during_df = max_anomalies_during.add_prefix("AMAX_")
```

```python
%%time
with multiprocessing.Pool(30) as p:
    res_during_min = p.map(summarize_event_min_anomalies_originals_duringevent, events_list)
```

```python
min_anomalies_during = pd.concat(res_during_min)
```

```python
min_anomalies_during_df = min_anomalies_during.add_prefix("AMIN_")
```

```python
%%time
with multiprocessing.Pool(30) as p:
    res_during_mean = p.map(summarize_event_mean_anomalies_originals_duringevent, events_list)
```

```python
mean_anomalies_during = pd.concat(res_during_mean)
```

```python
mean_anomalies_during_df = mean_anomalies_during.add_prefix("AMEAN_")
```

```python
clim_summary_anomalies_original = max_anomalies_during_df.join(min_anomalies_during_df).join(mean_anomalies_during_df)
```

```python
#clim_summary_anomalies_original = clim_summary_anomalies_original.add_suffix("_O")
```
